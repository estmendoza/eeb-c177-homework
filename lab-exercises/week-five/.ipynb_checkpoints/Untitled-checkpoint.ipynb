```python
import numpy as np
```

```python
data = np.loadtxt(fname='../week-four/inflammation-01.csv', delimiter=",")
```

```python
import matplotlib.pyplot as plt
```

```python
image = plt.imshow(data)
plt.show()
```

```python
ave_inflammation = np.mean(data, axis=0) 
#mean along rows=0 along collumns =1
ave_plot = plt.plot(ave_inflammation)
plt.xlabel('time (days)')
plt.ylabel('inflammation score')
plt.title('patient data')
plt.show()
```

```python
max_plot = plt.plot(np.max(data, axis=0))
#for each data, the maximum inflammatio score recorded
plt.show()
```

```python
min_plot = plt.plot(np.min(data, axis=0))
#for each data, the minimum inflammation score recorded
plt.show()
```

```python
fig = plt.figure(figsize=(10.0, 3.0))
#object that stores multiple plots

fig.suptitle('this is a title', va='center')

subplt1 = fig.add_subplot(1, 3, 1)
subplt2 = fig.add_subplot(1, 3, 2)
subplt3 = fig.add_subplot(1, 3, 3)
#3 placeholder variables (which position on the figure you want each to be)
subplt1.set_ylabel('average')
subplt1.plot(np.mean(data, axis=0))

subplt2.set_xlabel(('time (days)'))
subplt2.set_ylabel('max')
subplt2.plot(np.max(data, axis=0))

subplt3.set_ylabel('min')
subplt3.plot(np.min(data, axis=0))

fig.tight_layout()

plt.show()
```

```python
#get all even numbers from 1 to 100
evens = []
for ii in range(1,101):
#will stop at 100 bc thats how range works 
    if (ii % 2 == 0):
        evens.append(ii)
        
print(evens)
```

```python
#get all odd numbers from 1 to 100
odds = []
for ii in range(1,101):
#will stop at 100 bc thats how range works 
    if (ii % 2 == 1):
        odds.append(ii)
        
print(odds)
```

```python
#find squares for first 10 even numbers
squares_nums = [i * i for i in range(20) if i % 2 == 0]
print(squares_nums)
```

```python
#get all even numbers from 0 to 100
evens = [i for i in range(101) if i % 2 == 0]
print(evens)
```
