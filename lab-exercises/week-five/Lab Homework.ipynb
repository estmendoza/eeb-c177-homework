```python
import numpy as np
```

```python
data = np.loadtxt(fname='../week-four/inflammation-01.csv', delimiter=",")
```

```python
import matplotlib.pyplot as plt
```

```python
sd_inflammation = np.std(data, axis=0)

sd_plot = plt.plot(sd_inflammation)
plt.xlabel('time (days)')
plt.ylabel('standard deviation score')
plt.title('standard deviation plot')
plt.show()
```

```python
import matplotlib.gridspec as gridspec
gs = gridspec.GridSpec(3,1)

fig = plt.figure(figsize = (10,15))

fig.suptitle("patient inflammation data", fontsize=40)

subplt1 = fig.add_subplot(gs[0])
subplt2 = fig.add_subplot(gs[1])
subplt3 = fig.add_subplot(gs[2])

subplt1.set_title('\n', size=40)
subplt1.set_ylabel('average', size=40)
subplt1.plot(np.mean(data, axis=0))

subplt2.set_ylabel('max', size=40)
subplt2.plot(np.max(data, axis=0))

subplt3.set_xlabel('time (days)', size=40)
subplt3.set_ylabel('min', size=40)
subplt3.plot(np.min(data, axis=0))

fig.tight_layout()

plt.show()
```

```python
x = "Ucla" 

def reverse(x):
    str = ""
    for ii in x:
        str = ii + str
    return str

print(x)
print (reverse(x))
```

```python
vowels = ['a' , 'e' , 'i' ,'o' , 'u']

x = "Shawn is a helpful TA"
count = 0
for ii in x.lower():
   if ii in vowels:
    count += 1
print("vowels in first string: " + str(count))


x = "My brain is busted"
count = 0
for ii in x.lower():
   if ii in vowels:
    count += 1
print("vowels in second string: " + str(count))
```
